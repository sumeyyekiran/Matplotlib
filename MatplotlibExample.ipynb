```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()
r = np.linspace(0, 10, num=100)
df = pd.DataFrame({'r': r, 'slow': r, 'medium': 2 * r, 'fast': 4 * r})
df = pd.melt(df, id_vars=['r'], var_name='speed', value_name='theta')
g = sns.FacetGrid(df, col="speed", hue="speed",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False)
g.map(plt.scatter, "theta", "r")
```

```python
import numpy as np 

class GaussKernel:

    def __init__(self, kernel_size, sigma):
        self.kernel_size = kernel_size
        self.sigma = sigma # varyans olcusu
        self.kernel = self._createKernel()

    def _createKernel(self):
        l = (self.kernel_size - 1) // 2 # kernel boyutu 2*l+1 
        x = np.array(range(-l,l+1)) # bir -l, -l+1.....l-2,l-1,l cinsinde bir array toplamda kernel_size kadar elemani var
        xx, yy = np.meshgrid(x,x) # buradan kernel_size x kernel_size'lik bir koordinar sistemi elde ettik meshgrid ile
        kernel = np.exp(-1 *( np.power(xx,2) + np.power(yy,2) )/ 2.0 * np.power(self.sigma,2))  # 2 boyutlu gauss fonksiyonu
        return kernel / np.sum(kernel) ## sonucu normallestiriyoruz
    
    def getKernel(self):
        return self.kernel

g = GaussKernel(11,1.0)
kernel =  np.log10(g.getKernel()) # log alarak daha "smooth" bir kernel olusturduk

from matplotlib import pyplot as plt  
plt.imshow(kernel,interpolation='none')
plt.show()
```
